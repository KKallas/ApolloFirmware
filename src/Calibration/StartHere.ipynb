# Welcome to the Apollo Calibration Toolkit

The Apollo Calibration Toolkit is an essential resource for anyone looking to set up and calibrate DS Apollo LED cinema lamps. This comprehensive toolkit includes:

### Jupyter Lab Workbooks
- Interactive and user-friendly guides that walk you through the calibration process step-by-step.

### Detailed Drawings and Instructions
- Visual aids and clear instructions to assist in the assembly and calibration procedures.

### Bill of Materials (BOM)
- A complete list of all the components required for building the calibration setup.

### 3D Printing Files
- Custom-designed files for creating the mechanical parts necessary for the calibration workbench.

Designed with versatility in mind, this toolkit ensures compatibility with all major PC platforms, including **Windows**, **Mac OS**, and **Linux**. Whether you're a professional technician or an enthusiast, the Apollo Calibration Toolkit provides everything you need to initialize and calibrate your Apollo LED cinema lamps efficiently and accurately.

Explore the links below to access all the necessary tools and embark on a seamless calibration experience.


## Tools

* [Setup](Setup.ipynb) - Jupyter lab library install
* Burn Software - In Arduino flash the firmware to the ESP32 Controller
* [Lamp Intilize](InitLamp.ipynb) - Write Intial Values into FLASH memory
* [Lamp Calibration](TestCalibV2.ipynb) - Generate 4D LUT to calibrate the lamp to White Balance mode
* [Lamp Measure](2800-10000K_testV2.ipynb) - Measure and log all whitebalance brighntess values for QC
* [Set DMX](SetDmxAddr.ipynb) - Get and Set DMX offset for the serial connected lamp
